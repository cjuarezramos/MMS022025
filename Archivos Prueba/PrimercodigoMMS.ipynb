```python
# Módulos de python a utilizar
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Información del sistema, parámetros, condiciones iniciales.
B = 5 # Coeficiente de fricción kg/s
M = 80 # Masa del objeto en kg
v0 = 1 # Velocidad inicial m/s
```

```python
# Discretización del tiempo
t0 = 0.0
tf = 100.0
N = 10    # Número de pasos
dt = (tf - t0) / N
t = np.linspace(t0, tf, N + 1) # Vector de tiempo, linspace crea un vector de N+1 puntos equidistantes
```

```python
# solución de la ecuación diferencial analítica
v_analitica = v0 * np.exp(-B/M * t)
plt.plot(t, v_analitica, label='Solución analítica', color='blue')
```

```python
# Solución numérica por el método de Euler
v_euler = np.zeros(N + 1) # Vector de velocidades, inicializado en cero
v_euler[0] = v0 # Condición inicial
for i in range(N):
    v_euler[i + 1] = v_euler[i] - (B/M) * v_euler[i] * dt
# Graficar resultados
plt.plot(t, v_analitica, label='Solución analítica', color='blue')
plt.plot(t, v_euler, label='Solución numérica (Euler)', color='red')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.title('Comparación de soluciones analítica y numérica')
plt.legend()
```
