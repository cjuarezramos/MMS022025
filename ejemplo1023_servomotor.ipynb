```python
# Librerìas
import numpy as np
import matplotlib.pyplot as plt
import control as ct
```

```python
# Servomotor
K = 1  # Nm/V
B = 0.1   # Nms/rad
J = 0.01  # kgm^2
#Funciòn de transferencia
G = ct.tf([K/J], [1, B/J, K/J])
print(G)
```

```python
sol = ct.step_response(G)
plt.plot(sol.time, sol.outputs)
plt.title('Respuesta al Escalón del Servomotor')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (rad)')
plt.grid()
```

```python
# Lugar de las raíces
G = ct.tf([1], [J, B, 0])
print(G)
ct.root_locus(G)
plt.title('Lugar de las Raíces del Servomotor')
plt.grid()
```

```python
G = ct.tf([1,1],[1,0,0])
print(G)
ct.root_locus(G)
plt.title('Lugar de las Raíces del Servomotor con Polo y Cero')
plt.grid()
```

```python
sol = ct.step_response(G)
plt.plot(sol.time, sol.outputs)
plt.title('Respuesta al Escalón del Servomotor con Polo y Cero')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (rad)')
plt.grid()
```

```python
K = 100
T = ct.feedback(K*G, 1)
print(T)
sol = ct.step_response(T)
plt.plot(sol.time, sol.outputs)
plt.title('Respuesta al Escalón del Servomotor con Polo, Cero y Controlador Proporcional')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (rad)')
plt.grid()
```
